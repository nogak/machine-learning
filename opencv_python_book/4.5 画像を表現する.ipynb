```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img_bgr = cv2.imread('data/lena.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img_bgr)
plt.subplot(122)
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
# 
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)

plt.imshow(corners, cmap='gray')
```

```python
# SIFT (scale-invariant feature transform)
sift = cv2.xfeatures2d_SIFT_create()
kp = sift.detector()

import numpu as np
img_kp = np.zeros_like(img_bgr)
img_kp = cv2.drawKeypoints(img_rgb, kp, img_kp,
                           flags=cv2.DRAWPMATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_kp)
```
